```python
import pandas as pd
```

```python
df = pd.read_csv('Contest Data/contest-standings-9_3_20.csv')
df.head()
```

```python
df1 = df[['Rank', 'EntryName', 'Points', 'Lineup']]
df1.head()
```

```python
def sample_first_prows(data, perc=0.1):
    return data.head(int(len(data)*(perc)))

df_top10 = sample_first_prows(df1)
df_top10.head()
```

```python
print(len(df))
print(len(df_top10))
```
